## 方向导数
偏导数反映的是多元函数沿着**坐标轴**方向的变化率，方向导数本质上研究的是函数在**某点**处沿着**特定方向**的变化率问题。

例如y=y_0，f(x,y_0)得到是三维空间中，与x轴平行的一个平面

可以理解为偏导数是
1. $y=y_0$生成一个与x轴平行，垂直于y轴的$y=y_0$平面，同原函数例如$z=x^2+y^2$相交，得到一个曲线函数，求得该函数在曲线上各个点的斜率，但斜率线方向始终平行于x，垂直于y。
3. $x=x_0$生成一个与y轴平行，垂直于x轴的$x=x_0$平面，同原函数例如$z=x^2+y^2$相交，得到一个曲线函数，求得该函数在曲线上各个点的斜率，但斜率线方向始终平行于y，垂直于x。

但是若x，y同时变化，而非其中一个值固定，
首先考虑x，y同时变化到一个固定值即点$p(x_0,y_0)$那么在三维空间中，x，y构成的底座二维空间就会有一个$l$向量，该向量以$f(x=0,y=0,z=0)$为起点以$p(x_0,y_0)$为方向，同时不断增加$\Delta{x},\Delta{y}$。

该向量直线向z轴衍生，构成一个平面，并于原函数$z=f(x,y)$相交成为一个曲线，此时求改曲线的各个点斜率就是$l$方向导数。
1. 三维空间中，$l$向量是在二维${x,y}$空间中生成的，是$e_x,e_y$的线性组合。
2. 

$$\frac{\partial f}{\partial l}=\lim_{p \to 0} \frac{f(x_0+\Delta{x},y_0+\Delta{y})-f(x_0,y_0)}{p}$$
沿着x，y相互垂直坐标轴，生成的平面在点(x,yo)上与f(x,y)切面的切线斜率，而方向导数l是在空间中任意位置（由x，y轴e_x,e_y合力表达)

Example 1
Let $f(x,y)=x^2y$. (a) Find ∇f(3,2). (b) Find the derivative of f in the direction of (1,2) at the point (3,2).
Example 2
For the f of Example 1, find the directional derivative of f at the point (3,2) in the direction of (2,1).

```python
from sympy import * 
import math
import matplotlib.pyplot as plt 
from sympy.plotting import *
import numpy as np 
def cal_degree(vec):
    """计算二维向量对应角度
    :param vec:输入二维向量
    :returns: 返回向量对应角度
    """
    x=vec[0]
    y=vec[1]
    l=np.linalg.norm(vec)
    cosxy=x/l
    print("向量cos值：",cosxy)
    degree=math.degrees(math.acos(cosxy))
    return(degree)
def grad(exp,x0,y0,plot=False):
    """ 计算梯度
    :param exp: 输入公式
    :param x0: 初始x值
    :param y0: 初始y值
    :returns: 返回梯度向量,以及向量对应角度
    :raises keyError: raises an exception
    """
    x,y=symbols("x,y")
    exp=simplify(exp)
    dif_x=diff(exp,x)
    dif_y=diff(exp,y)
    r0=dif_x.evalf(subs={x:x0,y:y0})
    r1=dif_y.evalf(subs={x:x0,y:y0})
    grad_v=np.array([float(r0),float(r1)])
    if plot==True:
        fig,ax=plt.subplots()
        fig.set_size_inches(10,10)
        ax.grid()
        ax.arrow(0,0,grad_v[0],grad_v[1],head_width=0.1,color='green')
    degree=cal_degree(grad_v)
    return(grad_v,degree)
def direct_diff(exp,x0,y0,lx,ly):
    """在梯度基础上计算方向向量，方向向量=梯度向量.T@(cosa,sina)其中a为方向向量与x轴的角度
    :param exp:输入原函数公式
    :param x0: 初始x值
    :param y0: 初始y值
    :param lx:方向向量x值
    :param ly:方向向量y值
    :returns: 方向向量导数值，x0y0点的梯度向量，梯度角度，方向向量的方向角度
    """
    x,y=symbols("x,y")
    exp=simplify(exp)
    l_vec=np.array([lx,ly])
    grad_v,grad_degree=grad(exp,x0=x0,y0=y0)
    l_norm=np.linalg.norm(l_vec)
    cos_delta=lx/l_norm
    sin__delta=ly/l_norm
    delta_vec=np.array([cos_delta,sin__delta])
    l_degree=cal_degree(l_vec)
    rs=grad_v.T@delta_vec
    return(rs,grad_v,grad_degree,l_degree)
if __name__=='__main__':
    x,y=symbols("x,y")
    z=x**2*y
    grad_v=grad(z,x0=3,y0=2)
    print("梯度：",grad_v)
    rs,grad_v,grad_degree,l_degree=direct_diff(z,x0=3,y0=2,lx=1,ly=2)
    print("方向导数值",rs)
    print("方向导数角度：",l_degree)
    print("x0,y0点的梯度角度：",grad_degree)
    print("x0,y0点的梯度：",grad_v)

 
```

## 使用梯度上升法求全局最大值（求全局最小值）


有时候只是想求一个函数的全局最大值，而不是所有的局部和全局最大值。
使用梯度上升法（gradient ascent mehtod)是一种迭代求全局最大值的方法。其涉及大量计算，故适合程序化求解而非手工运算。

--- 
> 简单而言，一维x的梯度就是f(x)对x的维分，而多维（也称为多变量）向量x的梯度就是f(x)对x的所有元素的偏微分构成一个偏导数向量。
也就是说，一元函数除了通过二次导数方式求极值，也可以通过梯度法求得极值。

梯度法中通过对自变量构成的方向向量$\vec{l_{(x,y)}}$不断的进行改变，来找到函数$f(x,y)$中最小值。

由于根据方向导数与梯度关系可知，在点$M(x_0,y_0)$附近$\vec{l_{(x_0,y_0)}}=\nabla{f(x_0,y_0)} \dot [\cos(\sigma),\sin(\sigma)]$,其中$\sigma$为方向导数与x轴夹角

$\longrightarrow$

其中$[\cos(\sigma),\sin(\sigma)]=[\cos(\sigma) \dot{e_{(1,0)}},\cos(\alpha) \dot e_{(0,1)}]$转换为单位坐标构成的一个向量 

$\longrightarrow$

即$\vec{l_{(x_0,y_0)}}=\nabla{f(x_0,y_0)} \dot [\cos(\sigma),\sin(\sigma)]$ 两个向量的内积为向量1的模在向量2的投影的乘积

$\longrightarrow$

$\vec{l_{(x_0,y_0)}}=||\nabla{f(x_0,y_0)} || \times ||\dot ||[\cos(\sigma),\sin(\sigma)]|| \times \cos(\gamma)$，由于$||[\cos(\sigma),\sin(\sigma)]||=1$

$\longrightarrow$

$\vec{l_{(x_0,y_0)}}=||\nabla{f(x_0,y_0)} || \times 1 \times \cos(\gamma)$ 其中$\gamma$为$\nabla{f(x_0,y_0)}$与$\vec{l_{(x_0,y_0)}}$夹角

$\longrightarrow$

得到当该方向向量与梯度一致时，$\gamma=0 \rightarrow \cos(\gamma)=1$方向导数达到最大=$||\nabla{f(x_0,y_0)} ||$

$\longrightarrow$

原方程$f(x,y)$上升（下降）速率在该点最大。此时
 

基于以上，可以根据以下步骤找到函数最大值

梯度法的步骤：
1. 设置函数初始点$M(x_0,y_0)$,以及学习率$$，

2. 通过移动点M到新的点，来找到原函数f(x,y)上升下载最大方向。
  * $M_new=M_old+\phi \nabla{f(x_0,y_0)} $(最大上升速率)
  * $M_new=M_old-\phi \nabla{f(x_new,y_new)} $(最大下降速率)
> 其中$\phi \nabla{f(x_0,y_0)} $在偏导方向上移动的一个固定数值*当前点梯度值。

3. 重复步骤2，移动到新的点$M_{new(n+1)}=M_{old(n)}+\phi \nabla{f(x_0,y_0)}$ 

4. 判断如何跳出循环，有两种方法：
   * 设定迭代次数，如果达到就跳出循环
   * 迭代后$\phi \nabla{f(x_0,y_0)}$小于某个数值（极小值）：说明梯度模不断减小，新点与老点距离差异达到极小值，逐渐达到原函数顶点


e.g. 有如下损失函数：

$$y=7t^2-1120t+45850$$
通过梯度下降法求其最小值
1. 通过sympy求
2. 通过numpy求
设起点$M=(20)$学习率$\gamma=0.001$
* 首先使用指定迭代次数
* 比较新x值与老x值小于极小值

```python
#指定迭代次数求最小值
from sympy import * 
import math
import matplotlib.pyplot as plt 
from sympy.plotting import *
import numpy as np 
y,x,sigma,x_0,x_old,x_new=symbols("y,x,sigma,x_0,x_old,x_new")
sigma=0.01
y=7*x*x-1120*x+45850
plot(y,(x,-10000,10000))#由于数值
grad_x=diff(y,x)
pprint(grad_x)
x_0=20
x_old=0
x_new=0
for i in range(1,16):
    #计算新的点
    if i==1:
        x_old=20
    else:
        x_old=x_new
    x_new=x_old-sigma*grad_x
    pprint(x_new)
    x_new=x_old-sigma*grad_x.subs({x:x_old})

    print("迭代:",i)
    print("x_old:",x_old)
    print("x_new:",x_new)
print("最终点：",x_new)
print("将最终值代入原函数得到极小值：",y.subs({x:x_new}))
 
```

```python
# 通过比较新自变量与旧自变量差小于极小值，来判断是否跳出自变量更新循环
#指定迭代次数求最小值
from sympy import * 
import math
import matplotlib.pyplot as plt 
from sympy.plotting import *
import numpy as np 
y,x,sigma,x_0,x_old,x_new=symbols("y,x,sigma,x_0,x_old,x_new")
sigma=0.01
y=7*x*x-1120*x+45850
grad_x=diff(y,x)
pprint(grad_x)
x_0=20
x_old=0
x_new=0
min_value=5e-9
i=1
over=False
while over==False:
    #计算新的点
    if i==1:
        x_old=20
    else:
        x_old=x_new
    x_new=x_old-sigma*grad_x
    pprint(x_new)
    x_new=x_old-sigma*grad_x.subs({x:x_old})
    dev=abs(x_old-x_new)
    print(dev)
    if  dev>min_value:
        print("尚未达到极小值，继续迭代")
        over=False
    else:
        over=True
    print("迭代:",i)
    i=i+1
    
    print("x_old:",x_old)
    print("x_new:",x_new)
print("最终点：",x_new)
print("将最终值代入原函数得到极小值：",y.subs({x:x_new}))
 
```

实现多元方程的梯度下降
1. 实现通用的函数来求得极小值
2. 调整学习率来实现对比

```python
from sympy import * 
import math
#import matplotlib.pyplot as plt 
from sympy.plotting import *
import numpy as np 
def param_min(starPoint,sigma,min,exp):
    """_通过梯度下降法，基于梯度，找到原函数下降最大方向，并指定学习率，不断更新自变量，从而逼近梯度，直到自变量下降趋近于极小值_
    
    """
 
    x,y,x_new,y_new,x_old,y_old=symbols("x,y,x_new,y_new,x_old,y_old")
    exp=simplify(exp)
    dif_x=diff(exp,x)
    pprint(dif_x)
    dif_y=diff(exp,y)
    pprint(dif_y)
    x_old=starPoint[0]
    y_old=starPoint[1]
    x_new=x_old-sigma*dif_x.subs({x:x_old,y:y_old})
    y_new=y_old-sigma*dif_y.subs({x:x_old,y:y_old})
    i=1
    while abs(exp.evalf(subs={x:x_old,y:y_old})-exp.evalf(subs={x:x_new,y:y_new}))>min:
 
        print("尚未达到极小值，继续迭代")
        x_old=x_new
        y_old=y_new
        x_new=x_old-sigma*dif_x.subs({x:x_old,y:y_old})
        y_new=y_old-sigma*dif_y.subs({x:x_old,y:y_old})

        print("迭代:",i)

        print("x新值：",x_new)
        print("y新值：",y_new)
        i=i+1
    return(x_new,y_new,exp.evalf(subs={x:x_new,y:y_new}))
if __name__=="__main__":
    print("begin")
    x_0,y_0,x,y,x_new,y_new,x_old,y_old,sigma,min=symbols("x_0,y_0,x,y,x_new,y_new,x_old,y_old,sigma,min")
    exp=x**2+y**2
    
    startPoint=np.array([2,3])
    sigma=0.3
    min=7e-8
    x,y,res=param_min(startPoint,sigma,min,exp)

    print(x,y,res)
    startPoint=np.array([0,0])
    x,y=symbols("x,y")
    exp=x-y+x**2+2*x*y+y**2
    x,y,res=param_min(startPoint,sigma,min,exp)
    print(x,y,res)
```
